```python
import numpy as np

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# weights
w1, w2, w3, w4 = -0.15, 0.20, 0.25, 0.30
w5, w6, w7, w8 = 0.40, 0.45, 0.50, 0.26

# bias
b1, b2 = 0.5, 0.7

# inputs
x1, x2 = 0.05, 0.1  

# outputs of hidden layers
h1_input = w1 * x1 + w2 * x2 + b1
h1 = tanh(h1_input)


h2_input = w3 * x1 + w4 * x2 + b1
h2 = tanh(h2_input)

# outputs of ouputs layers
o1_input = w5 * h1 + w6 * h2 + b2
o1 = tanh(o1_input)

o2_input = w7 * h1 + w8 * h2 + b2
o2 = tanh(o2_input)

print("O1:", o1)
print("O2:", o2)


# visualization
import matplotlib.pyplot as plt
import networkx as nx
G = nx.DiGraph()
# Nodes
G.add_nodes_from(["i1", "i2", "h1", "h2", "o1", "o2"])

# Edges with weights
edges = [("i1", "h1", w1), ("i1", "h2", w3),
         ("i2", "h1", w2), ("i2", "h2", w4),
         ("h1", "o1", w5), ("h1", "o2", w7),
         ("h2", "o1", w6), ("h2", "o2", w8)]
G.add_weighted_edges_from(edges)

# Positioning
pos = {"i1": (-1, 1), "i2": (-1, -1), "h1": (0, 1), "h2": (0, -1), "o1": (1, 0.5), "o2": (1, -0.5)}

plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='black', node_size=2000, font_size=10)
labels = {(u, v): f'{d:.2f}' for u, v, d in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Simple Visualization")
plt.show()


```
